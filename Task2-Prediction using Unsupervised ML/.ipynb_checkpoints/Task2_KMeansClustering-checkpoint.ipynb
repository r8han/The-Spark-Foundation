```python
sns.heatmap(iris_data.corr(), annot=True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix
```

```python
SSE = []
for i in range(1,10):
    kmeans = KMeans(n_jobs = -1, n_clusters = i, init='k-means++')
    kmeans.fit(iris_data.iloc[:,[0,1,2,3]])
    SSE.append(kmeans.inertia_)
```

```python
df = pd.DataFrame({'Cluster':range(1,10), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(df['Cluster'], df['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia');
plt.title("'ELBOW METHOD TO DETERMINE OPTIMAL VALUE OF 'K'\n");
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 3, init='k-means++')
kmeans.fit(iris_data.iloc[:,[0,1,2,3]])
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
iris_data['cluster'] = kmeans.labels_

iris_data
```

```python
display(iris_data['cluster'].value_counts(), iris_data['Species'].value_counts())
```

```python
plt.figure(figsize=(10,5))
plt.scatter(iris_data['SepalLengthCm'], iris_data['SepalWidthCm'], c=iris_data.cluster)
plt.title('Predicted Clusters\n')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1] ,  s=200, c='red', label = 'Centroids')
plt.show()
```

```python
iris_data.loc[iris_data['Species']=='Iris-setosa']['cluster'].value_counts() ## the model has identified iris-setosa with 100% accuracy
```

```python
iris_data.loc[iris_data['Species']=='Iris-versicolor']['cluster'].value_counts()
```

```python
iris_data.loc[iris_data['Species']=='Iris-virginica']['cluster'].value_counts()
```

```python
iris_data['SpeciesEncoded'] = iris_data['Species'].apply(lambda x: 1 if x=='Iris-setosa' else 2 if x=='Iris-virginica' else 0)
iris_data
```

```python
print(classification_report(iris_data['SpeciesEncoded'],iris_data['cluster']))
```

```python
sns.heatmap(confusion_matrix(iris_data['SpeciesEncoded'],iris_data['cluster']), annot=True)
```
